```python
import pandas as pd
import glob
import os

# Đường dẫn tới thư mục chứa file CSV
folder_path = r"D:\finalNodeJs\xulydata\data product"   # <= sửa nếu khác

# Lấy danh sách file CSV trong thư mục
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

print("Found files:")
csv_files
```

```python
dfs = []

for file in csv_files:
    try:
        df = pd.read_csv(file, encoding="utf-8")
    except:
        df = pd.read_csv(file, encoding="utf-8-sig")
    dfs.append(df)

merged = pd.concat(dfs, ignore_index=True)
merged.drop_duplicates(inplace=True)

merged.head()
```

```python
output_file = os.path.join(folder_path, "merged_yame.csv")
merged.to_csv(output_file, index=False, encoding="utf-8-sig")

output_file
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đảm bảo hiển thị tiếng Việt ổn (thường DejaVu Sans là mặc định)
plt.rcParams["font.family"] = "DejaVu Sans"

df = pd.read_csv("yame_products_enriched.csv")

df.head()
```

```python
df.info()
```

```python
df.describe(numeric_only=True)
```

```python
category_counts = df["category"].value_counts()

plt.figure(figsize=(10, 5))
category_counts.plot(kind="bar")
plt.title("Số lượng sản phẩm theo danh mục (category)")
plt.xlabel("Danh mục")
plt.ylabel("Số sản phẩm")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
avg_price_by_cat = df.groupby("category")["price_sale"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
avg_price_by_cat.plot(kind="bar")
plt.title("Giá sale trung bình theo danh mục")
plt.xlabel("Danh mục")
plt.ylabel("Giá sale trung bình (VND)")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
